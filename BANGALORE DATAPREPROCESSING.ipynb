                            TO REPLACE THE DATE INTO STRING

```python
# Function to check if the date is before June print the string as it

def is_before_june(date_str):
    month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    try:
        day, month = date_str.split('-')
        return month_order.index(month) <=month_order.index('Jun')
    except ValueError:
        return False  

    # Apply the function to the "availability" column
df['availability'] = df['availability'].apply(lambda x: 'completed' if is_before_june(x) else x)


```

```python
# Function to check if the date is after June print the string 'in future'

def is_After_june(date_str):
    month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    try:
        day, month = date_str.split('-')
        return month_order.index(month) > month_order.index('Jun')
    except ValueError:
        return False
    
df['availability'] = df['availability'].apply(lambda x: 'In Future' if is_After_june(x) else x)

```

```python
#Extract the values alone from size ,replace the column name as 'bhk' and print the unique values


def extract_bhk(size):
    size = str(size).replace(' ', '')

    digits = ''.join(filter(str.isdigit, size))
    try:
        return int(digits[0])
    except (IndexError, ValueError):
        return np.nan
df['bhk'] = df['size'].apply(extract_bhk)
unique_bhk = df['bhk'].dropna().unique()
print("bhk",unique_bhk)
```

```python
df['total_sqft'].unique()
```

```python
#convert the 'total_sqft'range values to single value

def convert_range_to_single_value(range_str):
    if isinstance(range_str, str):
        try:
            if '-' in range_str:
                range_values = range_str.split('-')
                return (float(range_values[0].strip()) + float(range_values[1].strip())) / 2
            else:
                return float(range_str)
        except ValueError:
            return range_str
    else:
        return range_str

df['total_sqft'] = df['total_sqft'].apply(convert_range_to_single_value)

df['total_sqft']
```

```python
df['total_sqft'].unique()
```

```python
#Extract only the numeric value,remove the string

def extract_numeric_value(value):
    try:
        numeric_value = float(value.split('Sq. Meter')[0])
        return numeric_value
    except ValueError:
        # If the value cannot be converted, return NaN (or handle accordingly)
        return float('nan')

# Apply the function to the column after converting non-null values to strings
df['total_sqft'] = df['total_sqft'].astype(str).apply(extract_numeric_value)

df['total_sqft']
```
